```python
# Stats to collect
# Eventually add the chart to this
event_keys = []

week = []
event_type = []
district = []

q_auto_ave = []
q_score_ave = []
q_boulder_ave = []
q_high_goal_ave = []
q_low_goal_ave = []
q_foul_ave = []
q_cross_ave = []
q_tower_score_ave = []
q_win_margin_ave = []
q_win_score_ave = []

e_auto_ave = []
e_score_ave = []
e_boulder_ave = []
e_high_goal_ave = []
e_low_goal_ave = []
e_foul_ave = []
e_cross_ave = []
e_tower_score_ave = []
e_win_margin_ave = []
e_win_score_ave = []
```

```python
tbat.get_event_info('2016njbri').keys()
```

```python
for event, stats in all_stats.items():
    quals = stats['year_specific']['qual']
    elims = stats['year_specific']['playoff']
    
    # This code helped me identify issues with events missing data
    # print('Event: %s, Quals: %s, Elims: %s' % (event, type(quals), type(elims)))

    event_keys.append(event)
    
    event_info = tbat.get_event_info(event)
    week.append(event_info['week'])
    event_type.append(event_info['event_type_string'])
    if event_info['event_district_string'] == None:
        district.append('Regional')
    else:
        district.append(event_info['event_district_string'])

    q_auto_ave.append(quals['average_auto_score'])
    q_score_ave.append(quals['average_score'])
    q_boulder_ave.append(quals['average_boulder_score'])
    q_high_goal_ave.append(quals['average_high_goals'])
    q_low_goal_ave.append(quals['average_low_goals'])
    q_foul_ave.append(quals['average_foul_score'])
    q_cross_ave.append(quals['average_crossing_score'])
    q_tower_score_ave.append(quals['average_tower_score'])
    q_win_margin_ave.append(quals['average_win_margin'])
    q_win_score_ave.append(quals['average_win_score'])

    e_auto_ave.append(elims['average_auto_score'])
    e_score_ave.append(elims['average_score'])
    e_boulder_ave.append(elims['average_boulder_score'])
    e_high_goal_ave.append(elims['average_high_goals'])
    e_low_goal_ave.append(elims['average_low_goals'])
    e_foul_ave.append(elims['average_foul_score'])
    e_cross_ave.append(elims['average_crossing_score'])
    e_tower_score_ave.append(elims['average_tower_score'])
    e_win_margin_ave.append(elims['average_win_margin'])
    e_win_score_ave.append(elims['average_win_score'])
```

```python
df = pd.DataFrame({
    'Auto-q': q_auto_ave,
    'Score-q': q_score_ave,
    'BoulderScore-q': q_boulder_ave,
    'HighGoals-q': q_high_goal_ave,
    'LowGoals-q': q_low_goal_ave,
    'FoulScore-q': q_foul_ave,
    'CrossScore-q': q_cross_ave,
    'TowerScore-q': q_tower_score_ave,
    'WinMargin-q': q_win_margin_ave,
    'WinScore-q': q_win_score_ave,

    'Auto-e': e_auto_ave,
    'Score-e': e_score_ave,
    'BoulderScore-e': e_boulder_ave,
    'HighGoals-e': e_high_goal_ave,
    'LowGoals-e': e_low_goal_ave,
    'FoulScore-e': e_foul_ave,
    'CrossScore-e': e_cross_ave,
    'TowerScore-e': e_tower_score_ave,
    'WinMargin-e': e_win_margin_ave,
    'WinScore-e': e_win_score_ave,
    
    'Week': week,
    'EventType': event_type,
    'District': district
    }, index=event_keys)
```
